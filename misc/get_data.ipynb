```python
import os
from pathlib import Path
import csv
```

```python
root = Path('/projects/illinois/eng/cs/chackoge/lanne2/Datasets/clustering/clustered_subnetworks/104_sbm/')
out_root = Path('data/networks/orig/')

for network_dir in root.iterdir():
    if not network_dir.is_dir():
        print(f'{network_dir} is not a directory')
        continue

    network_id = network_dir.name

    cluster_file = network_dir / 'sbm' / 'com_without_o.dat'

    if not cluster_file.exists():
        print(f'{cluster_file} does not exist')
        continue

    out_dir = out_root / 'sbm' / network_id / 'sbm'
    # if out_dir.exists():
    #     print(f'{out_dir} already exists')
    #     continue
    out_dir.mkdir(parents=True, exist_ok=True)

    os.system(f'cp {out_root}/sbm_old/{network_id}/sbm/edge.dat {out_dir}/edge.dat')
    os.system(f'cp {cluster_file} {out_dir}/com.dat')
    os.system(f'cp {out_root}/sbm_old/{network_id}/sbm/source.txt {out_dir}/source.txt')
```

```python
root = Path('data/all_empirical_networks_cpp_cc/errors')

for network_dir in root.iterdir():
    if not network_dir.is_dir():
        continue

    log_file = network_dir / 'run_cpp_cc.err'
    if not log_file.exists():
        print(f'No log file for {network_dir}')
        continue

    with open(log_file) as f:
        lines = f.readlines()
        for line in lines:
            if 'Command being timed:' not in line:
                continue

            command = line.strip().split('Command being timed: ')[1]
            args = command.split(' ')
            # Take the argument of the flag '--existing-clustering'
            existing_clustering = args[args.index('--existing-clustering') + 1]
            if 'dwfeng2' not in existing_clustering:
                continue
            if 'degcor' not in existing_clustering:
                continue
            print(f'{network_dir.name}: {existing_clustering}')
```

# Get network/clustering

```python
old_root = '/projects/engrit/illinois-insper/chacko_warnow/sbm_eval/dwfeng2/sbm-membership/'
new_root = 'data/sbm-membership/'

network_root = Path('data/leiden_clustering/110-full-networks')
err_root = Path('data/all_empirical_networks_cpp_cc/errors/')
root = Path('data/sbm-membership/')
output_root = Path('data/networks/orig/sbm')
resolution_id = 'sbm'

for network_dir in err_root.iterdir():
    if not network_dir.is_dir():
        continue

    # Get network id
    network_id = network_dir.name

    # Get network file
    network_fp = network_root / f'{network_id}-clean.tsv'

    log_file = network_dir / 'run_cpp_cc.err'
    if not log_file.exists():
        print(f'No log file for {network_dir}')
        continue

    with open(log_file) as f:
        lines = f.readlines()
        existing_clustering = None
        for line in lines:
            if 'Command being timed:' not in line:
                continue
            else:
                command = line.strip().split('Command being timed: ')[1]
                args = command.split(' ')
                existing_clustering = args[args.index('--existing-clustering') + 1]
                break
        if existing_clustering is None:
            print(f'No existing clustering for {network_dir.name}')
    

    if old_root not in existing_clustering:
        print(f'Not from dwfeng2: {network_dir.name}')
        continue

    existing_clustering_fp = Path(existing_clustering.replace(old_root, new_root))

    # Get clustering file
    clustering_fp = existing_clustering_fp

    #  Get output directory
    output_dir = output_root / network_id / resolution_id

    # Create output directory
    output_dir.mkdir(parents=True, exist_ok=True)

    # Copy files
    os.system(f'cp {clustering_fp} {output_dir}/com.dat')
    os.system(f'cp {network_fp} {output_dir}/edge.dat')

    # Write the source of the clustering
    with open(output_dir / 'source.txt', 'w') as f:
        f.write(existing_clustering)

    # # Read clustering as CSV and write as TSV
    # with open(clustering_fp, 'r') as f:
    #     csv_reader = csv.reader(f, delimiter=' ')
    #     with open(output_dir / 'com.dat', 'w') as f_out:
    #         tsv_writer = csv.writer(f_out, delimiter='\t')
    #         for row in csv_reader:
    #             tsv_writer.writerow(row)
```

```python
network_root = Path('data/leiden_clustering/110-full-networks')
root = Path('data/all_empirical_networks_cpp_cc/output')
output_root = Path('data/networks/orig/sbm_cc')
resolution_id = 'sbm'

for network_dir in root.iterdir():
    if not network_dir.is_dir():
        continue

    # Get network id
    network_id = network_dir.name

    # Get network file
    network_fp = network_root / f'{network_id}-clean.tsv'

    # Get clustering file
    clustering_fp = network_dir / 'cpp_cc.clustering'

    #  Get output directory
    output_dir = output_root / network_id / resolution_id

    # Create output directory
    output_dir.mkdir(parents=True, exist_ok=True)

    # Copy files
    # os.system(f'cp {clustering_fp} {output_dir}/com.dat')
    os.system(f'cp {network_fp} {output_dir}/edge.dat')

    # Read clustering as CSV and write as TSV
    with open(clustering_fp, 'r') as f:
        csv_reader = csv.reader(f, delimiter=' ')
        with open(output_dir / 'com.dat', 'w') as f_out:
            tsv_writer = csv.writer(f_out, delimiter='\t')
            for row in csv_reader:
                tsv_writer.writerow(row)
```

```python
network_root = Path('data/leiden_clustering/110-full-networks')
root = Path('data/all_empirical_networks_cpp_cc/output')
output_root = Path('data/networks/orig/sbm_wcc')
resolution_id = 'sbm'

for network_dir in root.iterdir():
    if not network_dir.is_dir():
        continue

    # Get network id
    network_id = network_dir.name

    # Get network file
    network_fp = network_root / f'{network_id}-clean.tsv'

    # Get clustering file
    clustering_fp = network_dir / 'cpp_wcc.clustering'

    #  Get output directory
    output_dir = output_root / network_id / resolution_id

    # Create output directory
    output_dir.mkdir(parents=True, exist_ok=True)

    # Copy files
    # os.system(f'cp {clustering_fp} {output_dir}/com.dat')
    os.system(f'cp {network_fp} {output_dir}/edge.dat')

    # Read clustering as CSV and write as TSV
    if not clustering_fp.exists():
        print(f'Clustering file does not exist: {clustering_fp}')
        continue
    
    with open(clustering_fp, 'r') as f:
        csv_reader = csv.reader(f, delimiter=' ')
        with open(output_dir / 'com.dat', 'w') as f_out:
            tsv_writer = csv.writer(f_out, delimiter='\t')
            for row in csv_reader:
                tsv_writer.writerow(row)
```

# Get statistics

```python
inp_root = Path('data/SBM_FullNetworks/FullNetwork_RECCS_OS1_SBM_104_Evaluation/RECCSv1/OS1')
out_root = Path('data/stats/RECCSv1_OS1/')

for network_dir in inp_root.iterdir():
    # Get network id
    network_id = network_dir.name

    for clustering_dir in network_dir.iterdir():
        # Get clustering id
        clustering_id = clustering_dir.name
        if clustering_id in ['0.001', '0.01', '0.1']:
            resolution_id = f'leiden{clustering_id[1:]}'
            clustering_id = 'leiden_cpm_nofiltcm'
        elif clustering_id == 'k10':
            resolution_id = 'k10'
            clustering_id = 'ikc_nofiltcm'
        elif clustering_id == 'mod':
            resolution_id = 'leidenmod'
            clustering_id = 'leiden_mod_nofiltcm'
        elif clustering_id == 'infomap':
            resolution_id = 'infomap'
            clustering_id = 'infomap_nofiltcm'
        elif clustering_id in ['sbm', 'sbm_cc', 'sbm_wcc']:
            resolution_id = 'sbm'
        else:
            raise ValueError(f'Unknown clustering id: {clustering_id}')

        # Get output directory
        output_dir = out_root / clustering_id / network_id / resolution_id

        # Create output directory
        output_dir.mkdir(parents=True, exist_ok=True)

        # Copy files
        os.system(f'cp -r {clustering_dir} {output_dir}/0/')
```
